### 3. Data Preparation

- Dropping columns and rows
- Duplicates
- Rename columns
- Feature creation
- Missing Values

```python
# Drop unecessary columns
columns_drop = ['ID', 'ICO Number', 'In-Country Partner', 'Certification Address', 'Certification Contact']
data.drop(columns=columns_drop , axis =1 , inplace = True)
data.columns
```

```python
# Rename columns to all lowercase and no spaces
data.columns = data.columns.str.lower()
data.columns = data.columns.str.replace(" ", "_")
data.columns
```

```python
# Check for duplicates
print(f"Dataframe contains {data.duplicated().sum()} duplicate rows")
```

```python
# Change our 'expiration' column to a datetime dtype
data['expiration'] = pd.to_datetime(data['expiration'])
```

```python
# Look for missing values.
data.isna().sum()
```

```python
# Fill in Missing values
data = data.fillna('N/A')
data.isna().sum()
```

Changes:
- The following rows were dropped: 'ID', 'ICO Number', 'In-Country Partner', 'Certification Address', 'Certification Contact'
- All column names were changed to lower case and all spaces were replaced with an underscore
- No duplicates were found
- Imputed the value 'N/A' for any column with missing values.

```python
def dist_analysis(column, bins):
    """
    The following function creates a distribution plot for a specified feature with a specified number of bins
    """
    fig , ax = plt.subplots(figsize = (7,7))
    sns.distplot(data[column] , ax= ax)
    ax.set_title =f'{column} Score'.upper()
    ax.set_xlabel(f"{column} Score".upper())
    plt.show()
```
